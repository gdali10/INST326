# Searching
This small program allows the users to see a report for a specific country that they enter. The users has 4 options:

Option 1: Displaying all information and graphs.

Option 2: Displaying brand, country, and stars.

Option 3: Displaying brand, variety, style, and country.

Option 4: Write to csv file (Please remember to enter .csv at the end)

Option 5: Prediction markets.
If the temperature is below or 20 degree Celsius, we predict the customers will buy ramen.
If the temperature is above 20 degree Celsius, we predict the customers will not buy ramen.

```python
# Ka Wai - function1 and function2
def function1():
    ask = input("Please enter any country: ")
    c = rf[rf.Country == ask]
    df = pd.DataFrame(c)
    print(df.to_string())
    #style plot
    style = c.Style.value_counts()
    style.plot.pie(autopct="%1.0f%%")
    plt.axis("tight")
    plt.title("STYLES OF RAMEN IN " + ask.upper())
    plt.show()
    #brand plot
    brand = c.Brand.value_counts().head(10)
    brand.plot.barh(color=["steelblue", "orange", "green", "red", "purple", "brown", "pink", "gray", "olive", "cyan"])
    plt.title("TOP 10 BRANDS OF RAMEN IN " + ask.upper())
    plt.show()
    #variety plot
    variety = c.Variety.value_counts().head(10)
    variety.plot.barh(color=["steelblue", "orange", "green", "red", "purple", "brown", "pink", "gray", "olive", "cyan"])
    plt.title("TOP 10 FLAVOR OF RAMEN IN " + ask.upper())
    plt.show()
def function2():
    ask = input("Please enter any country: ")
    c = rf[rf.Country == ask]
    df = pd.DataFrame(c)
    df = df[["Brand", "Country", "Stars"]]
    print(df.to_string())
# Gaurav - function3 and function4
def function3():
    ask = input("Please enter any country: ")
    c = rf[rf.Country == ask]
    df = pd.DataFrame(c)
    df = df[["Brand", "Variety", "Style", "Country"]]
    print(df.to_string())
def function4():
    ask = input("Please enter any country: ")
    c = rf[rf.Country == ask]
    out = input("Enter the name of the file:")
    df = pd.DataFrame(c)
    df.set_index("Review #", inplace=True)
    df.to_csv(out)
# Thinh - function5
def function5():
    woeid = 0
    t = 0
    #Finding woeid
    query = input("Enter a city: ").strip().replace(" ", "+")
    urlBase = "https://www.metaweather.com/api/location/"
    urlSearch = urlBase + "search?query=" + query
    uh = urllib.request.urlopen(urlSearch)
    data = uh.read().decode()
    js = json.loads(data)
    for k in js:
        woeid = k["woeid"]
        print("woeid: " + str(woeid))       
    #Finding date    
    date = input("Please enter the date (Format for the date: yyyy/mm/dd): ")
    urlRetrieve = urlBase + str(woeid) + "/" + str(date) + "/"
    uhRetrieve = urllib.request.urlopen(urlRetrieve)
    data2 = uhRetrieve.read().decode()
    js2 = json.loads(data2)
    for k in js2:
        w = k["weather_state_name"]
        t = k["the_temp"]
    print("The temperature on " + str(date) + " is: " + str(t) + " degree Celsius. The weather is " + str(w))
    if t <= 20:
        print("The customers will buy ramen")
    elif t > 20:
        print("The customers will not buy ramen")

if __name__ == "__main__":
    cy = input(" 1: Displaying all infomation and graphs.\n 2: Displaying brand, country, and stars.\n 3: Displaying brand, variety, style, and country.\n 4: Write to csv file.\n 5: Prediction markets.\n Your choice is:  ")
    if cy == "1":
        function1()
    elif cy == "2":
        function2()
    elif cy == "3":
        function3()
    elif cy == "4":
        function4()
    elif cy == "5":
        function5()
```

# Object-oriented programming

```python
# Yuming
class Ramen_dataset:
    """
    A class that allow user to change csv file

    Attributes:
        file (string): a file name that have the ramen dataset
    """
    def __init__(self, file):
        self.file = file
        self.filedf = pd.read_csv(file)
        self.last_review = int(self.filedf['Review #'].tail(1).item())
        
    def remove_review(self, review_number):
    """
    Remove a review

    Args:
        review_number (real number): the review that want to remove

    Returns:
        (string): wheather the review is successfully removed
    """
        if review_number < 1 or review_number > self.last_review:
            return "Invaild review number"
        else:
            self.filedf = self.filedf[self.filedf['Review #'] != review_number]
            temp = []
            temp_index = []
            for i in range(1, self.last_review):
                temp.append(i)
                temp_index.append(i-1)
            self.filedf['Review #'] = temp
            self.filedf.reset_index(drop=True, inplace=True)
            self.last_review -= 1
            self.save_csv(self.file)
            return "Successfully removed Review # " + str(review_number)

    def add_review(self, brand, variety, style, country, stars, top = np.NaN):
    """
    Add a review

    Args:
        brand (string): the brand name
        variety (string): the falovr of the ramen
        style (string): the style of the ramen
        country (string): the country of the ramen
        stars (integer): the rating of the ramen
        top (string): whether the ramen is in top 10
 
    Returns:
        (dataframe): the last row of the dataframe
    """
        self.last_review = int(self.last_review + 1)
        self.filedf = self.filedf.append(pd.Series([self.last_review, brand, variety, style, country, stars, top],
                                                   index=self.filedf.columns ), ignore_index=True)
        self.save_csv(self.file)
        return self.filedf.tail(1)
    
    def modify_review(self, review_num, brand=None, variety=None, style=None, country=None, stars=None, top=None):
    """
    Modify a review

    Args:
        review_num (integer): the review number
        brand (string): the brand name
        variety (string): the falovr of the ramen
        style (string): the style of the ramen
        country (string): the country of the ramen
        stars (integer): the rating of the ramen
        top (string): whether the ramen is in top 10
 
    Returns:
        (dataframe): display the row
    """
        if review_num < 1 or review_num > self.last_review:
            return "Invaild review number"
        else:
            review_num -= 1
            if brand != None:
                self.filedf.at[review_num, 'Brand'] = brand
            if variety != None:
                self.filedf.at[review_num, 'Variety'] = variety
            if style != None:
                self.filedf.at[review_num, 'Style'] = style
            if country != None:
                self.filedf.at[review_num, 'Country'] = country
            if stars != None:
                self.filedf.at[review_num, 'Stars'] = stars
            if top != None:
                self.filedf.at[review_num, 'Top Ten'] = top
            self.save_csv(self.file)
            return self.display_row(review_num+1)
    
    def display_row(self, review_num):
    """
    Display a review

    Args:
        review_num (integer): the review number
 
    Returns:
        (dataframe): a row of review
    """
        return self.filedf[self.filedf['Review #'] == review_num]
    
    def save_csv(self, filename):
    """
    Save the dataframe to a csv

    Args:
        filename (string): a file to overwrite
    """
        self.filedf.to_csv(filename, index = False)
```
